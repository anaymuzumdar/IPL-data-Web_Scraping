```python
# importing libraries

import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# assigning empty list to store the values for each column 

player = []
match = []
innings = []
notout = []
runs = []
highestscore = []
average = []
ballsfaced = []
strikerate = []
hundreds = []
fiftys = []
fours = []
sixes = []
```

```python
# (divided the loop in 2 parts becoz of long running time)
# Part 1



for i in range(2008,2021):
    page = requests.get('https://www.iplt20.com/stats/{}/most-runs'.format(i))
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    table = soup.find('table', {'class':'table table--scroll-on-tablet top-players'})
    
   
    
# for loop used for traversing the data/rows    
    for row in table.select('tr'):
        cols = row.select('td')
        
        playername = row.find('td', {'class':'top-players__freeze top-players__player'})
        mat = row.find('td', {'class':'top-players__m top-players__padded'})
        inn = row.find('td', {'class':'top-players__inns'})
        no = row.find('td', {'class':'top-players__no'})
        rns = row.find('td', {'class':'top-players__r is-active'})
        hs = row.find('td', {'class':'top-players__hs'})
        avg = row.find('td', {'class':'top-players__a'})
        
        # appending the respective values in their respective list/column header without html tags
        var = 0
        if playername is None:
            var=0
        else:
            player.append(playername.text)
        if mat is None:
            var=0
        else:
            match.append(mat.text)
        if inn is None:
            var=0
        else:
            innings.append(inn.text)
        if no is None:
            var=0
        else:
            notout.append(no.text)
        if rns is None:
            var=0
        else:
            runs.append(rns.text)
        if hs is None:
            var=0
        else:
            highestscore.append(hs.text)
        if avg is None:
            var=0
        else:
            average.append(avg.text)
```

```python

# Part 2



for i in range(2008,2021):
    page = requests.get('https://www.iplt20.com/stats/{}/most-runs'.format(i))
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    table = soup.find('table', {'class':'table table--scroll-on-tablet top-players'})
    
    
# for loop used for traversing the data/rows    
    for row in table.select('tr'):
        cols = row.select('td')
        
        
        bf = row.find('td', {'class':'top-players__b'})
        sr = row.find('td', {'class':'top-players__sr'})
        hdd = row.find('td', {'class':'top-players__100s'})
        fft = row.find('td', {'class':'top-players__50s'})
        frs = row.find('td', {'class':'top-players__4s'})
        sxs = row.find('td', {'class':'top-players__6s'})
        # appending the respective values in their respective list/column header without html tags
        var = 0
       
        if bf is None:
            var=0
        else:
            ballsfaced.append(bf.text)
        if sr is None:
            var=0
        else:
            strikerate.append(sr.text)
        if hdd is None:
            var=0
        else:
            hundreds.append(hdd.text)
        if fft is None:
            var=0
        else:
            fiftys.append(fft.text)
        if frs is None:
            var=0
        else:
            fours.append(frs.text)
        if sxs is None:
            var=0
        else:
            sixes.append(sxs.text)
    
        
```

```python
# getting the season column desired output

season = []
for i in range(2008,2021):
    page = requests.get('https://www.iplt20.com/stats/{}/most-runs'.format(i))
    for j in range(100):
        season.append(i)
# print(season)
len(season)
```

```python
# creating a dataframe

IPL_Batsman_df = pd.DataFrame({ 
                            'Player':player, 
                            'Match':match, 
                            'Innings':innings, 
                            'Not_Out':notout,
                            'Runs':runs, 
                            'Highest_Score':highestscore,
                            'Average':average,'Balls_Faced':ballsfaced,
                            'Strike_Rate':strikerate,
                            '100s':hundreds,
                            '50s':fiftys,
                            '4s':fours,
                            '6s':sixes,
                            'Season':season})
IPL_Batsman_df.tail()
IPL_Batsman_df.head()
```

```python
# calculating the sum of runs scored by boundaries and singles/doubles in the dataframe

Total_Runs = IPL_Batsman_df['Runs'].sum()
Total_4s = (IPL_Batsman_df['4s'].sum())*4
Total_6s = (IPL_Batsman_df['6s'].sum())*6
Sing_Doub = Total_Runs - (Total_4s + Total_6s) 

print('Total runs in all the season :', Total_Runs)
print('Total runs scored by 4s in all season :', Total_4s)
print('Total runs scored by 6s in all season :', Total_6s)
print('Total runs scored using singles and doubles in all season :', Sing_Doub)
```
